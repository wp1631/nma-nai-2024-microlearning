```python
from glob import glob
import pandas as pd
```

```python
csv_files = glob("*.csv")
```

```python
pd.read_csv(csv_files[0])
```

```python
def new_read_csv(name:str):
    df = pd.read_csv(name)
    df.loc[40,"model"] = df.loc[0,"model"]
    df.loc[40,"seed"] = df.loc[0,"seed"]
    df.loc[40,"epoch"] = 41
    df.loc[:,"seed"] = df.loc[:,"seed"].astype(int)
    df.loc[:,"epoch"] = df.loc[:,"epoch"].astype(int)-1
    return df
```

```python
file_dfs = (
    new_read_csv(item) 
    for item in csv_files
)
all_df = pd.concat(file_dfs, axis=0)
```

```python
all_df.to_csv("all_data.dat")
```

```python
from sklearn.manifold import MDS
X = all_df.iloc[:,4:]
embedding = MDS(n_components=2, normalized_stress='auto')
X_transformed = embedding.fit_transform(X)
X_transformed.shape
```

```python
import matplotlib.pyplot as plt
plt.scatter(X_transformed[:,0], X_transformed[:,1])
```

```python
transformed_df = all_df.iloc[:,1:6]
```

```python
transformed_df
```

```python
transformed_df.iloc[:,3:] = X_transformed
```

```python
transformed_df
```

```python
transformed_df["seed"] = transformed_df["seed"].astype(int)
transformed_df["epoch"] = transformed_df["epoch"].astype(int)
```

```python
transformed_df.to_csv("transformed_data.dat")
```

```python
transformed_df
```

```python
import numpy as np
algorithms = ["perturb", "kolepoll", "feedback", "node_perturb", "hebb", "backprop"]
seeds = [12345,9999,42]
colormapper = {
    "perturb": "g", 
    "kolepoll": "b", 
    "feedback": "r", 
    "node_perturb": "m", 
    "hebb": "c", 
    "backprop": "k"
}
synbolmapper = {
    "perturb": "o", 
    "kolepoll": "v", 
    "feedback": ">", 
    "node_perturb": "^", 
    "hebb": "<", 
    "backprop": "*"
}

fig, ax = plt.subplots(figsize=(10,8), dpi=300)
ax.set_axis_off()

for algorithm in algorithms:
    for seed in seeds:
        data = transformed_df[np.logical_and(transformed_df["seed"]==seed, transformed_df["model"]==algorithm)]
        ax.plot(data["0"], data["1"], markevery=5, c=colormapper[algorithm])
        ax.scatter(data["0"][::5], data["1"][::5], label=algorithm, marker=synbolmapper[algorithm], s=10*data["epoch"][::5], c=colormapper[algorithm])
ax.scatter(data["0"][0], data["1"][0], label="Start", marker="X", s=200, c="orange", alpha=1)

handles, labels = plt.gca().get_legend_handles_labels()
# Filter out unwanted labels (e.g., remove 'Linear')
filtered_handles = [h for h, l in zip(handles[::3], labels[::3])]
filtered_labels = [l for l in labels][::3]

# Set Legend with Filtered Handles and Labels
plt.legend(filtered_handles, filtered_labels, loc='best', fontsize='xx-large')
# plt.xlim(-10,20)
# plt.ylim(-20,10)
```
